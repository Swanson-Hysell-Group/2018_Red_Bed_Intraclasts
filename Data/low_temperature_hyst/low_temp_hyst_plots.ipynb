```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import matplotlib.colors as colors
```

```python
viridis = cm = plt.get_cmap('viridis') 
cNorm  = matplotlib.colors.Normalize(vmin=50, vmax=300)
scalarMap = matplotlib.cm.ScalarMappable(norm=cNorm, cmap='viridis')

color_50 = scalarMap.to_rgba(50)
color_50_hex = colors.rgb2hex(color_50)
color_100 = scalarMap.to_rgba(100)
color_100_hex = colors.rgb2hex(color_100)
color_150 = scalarMap.to_rgba(150)
color_150_hex = colors.rgb2hex(color_150)
color_200 = scalarMap.to_rgba(200)
color_200_hex = colors.rgb2hex(color_200)
color_250 = scalarMap.to_rgba(250)
color_250_hex = colors.rgb2hex(color_250)
color_300 = scalarMap.to_rgba(300)
color_300_hex = colors.rgb2hex(color_300)
```

```python
BRIC_26_loops = pd.read_csv('BRIC_26_lowt_loops.csv')

BRIC_26_loop_300 = BRIC_26_loops[BRIC_26_loops['Temp (K)']==300]
BRIC_26_loop_250 = BRIC_26_loops[BRIC_26_loops['Temp (K)']==250]
BRIC_26_loop_200 = BRIC_26_loops[BRIC_26_loops['Temp (K)']==200]
BRIC_26_loop_150 = BRIC_26_loops[BRIC_26_loops['Temp (K)']==150]
BRIC_26_loop_100 = BRIC_26_loops[BRIC_26_loops['Temp (K)']==100]
BRIC_26_loop_50 = BRIC_26_loops[BRIC_26_loops['Temp (K)']==50]
BRIC_26_loop_10 = BRIC_26_loops[BRIC_26_loops['Temp (K)']==10]

plt.plot(BRIC_26_loop_300['X.2'],BRIC_26_loop_300['Mf'],color=color_300_hex,label='300 K')
plt.plot(BRIC_26_loop_250['X.2'],BRIC_26_loop_250['Mf'],color=color_250_hex,label='250 K')
plt.plot(BRIC_26_loop_200['X.2'],BRIC_26_loop_200['Mf'],color=color_200_hex,label='200 K')
plt.plot(BRIC_26_loop_150['X.2'],BRIC_26_loop_150['Mf'],color=color_150_hex,label='150 K')
plt.plot(BRIC_26_loop_100['X.2'],BRIC_26_loop_100['Mf'],color=color_100_hex,label='100 K')
plt.plot(BRIC_26_loop_50['X.2'],BRIC_26_loop_50['Mf'],color=color_50_hex,label='50 K')
plt.hlines(0,-5,5,zorder=-1)
plt.xlim(-5,5)
ymin, ymax = plt.ylim()
plt.vlines(0,ymin,ymax,zorder=-1)
plt.ylim(ymin, ymax)
plt.legend()
plt.xlabel('applied field (T)')
plt.ylabel('magnetization (Am$^2$/kg)')
plt.text(4.5,ymin*0.9,'BRIC 26',horizontalalignment='right',verticalalignment='bottom',fontweight='bold',fontsize=18)
plt.show()
```
