```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import matplotlib.colors as colors
```

```python
viridis = cm = plt.get_cmap('viridis') 
cNorm  = matplotlib.colors.Normalize(vmin=50, vmax=300)
scalarMap = matplotlib.cm.ScalarMappable(norm=cNorm, cmap='viridis')

color_50 = scalarMap.to_rgba(50)
color_50_hex = colors.rgb2hex(color_50)
color_100 = scalarMap.to_rgba(100)
color_100_hex = colors.rgb2hex(color_100)
color_150 = scalarMap.to_rgba(150)
color_150_hex = colors.rgb2hex(color_150)
color_200 = scalarMap.to_rgba(200)
color_200_hex = colors.rgb2hex(color_200)
color_250 = scalarMap.to_rgba(250)
color_250_hex = colors.rgb2hex(color_250)
color_300 = scalarMap.to_rgba(300)
color_300_hex = colors.rgb2hex(color_300)
```
