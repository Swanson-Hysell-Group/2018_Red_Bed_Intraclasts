```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import matplotlib.colors as colors
```

```python
BRIC_20c_raw = pd.read_csv('BRIC_20c_ac_raw.csv')
BRIC_20c_modeled = pd.read_csv('BRIC_20c_ac_modeled.csv')
BRIC_31c_raw = pd.read_csv('BRIC_31c_ac_raw.csv')
BRIC_31c_modeled = pd.read_csv('BRIC_31c_ac_modeled.csv')
BRIC_22c_raw = pd.read_csv('BRIC_22c_ac_raw.csv')
BRIC_22c_modeled = pd.read_csv('BRIC_22c_ac_modeled.csv')
BRIC_26c_raw = pd.read_csv('BRIC_26c_ac_raw.csv')
BRIC_26c_modeled = pd.read_csv('BRIC_26c_ac_modeled.csv')
```

```python
viridis = cm = plt.get_cmap('viridis') 
cNorm  = matplotlib.colors.LogNorm(vmin=1, vmax=1000)
# scalarMap = matplotlib.cm.ScalarMappable(norm=cNorm, cmap='viridis_r')
scalarMap = matplotlib.cm.ScalarMappable(norm=cNorm, cmap='viridis')

# for n in poles.index:
#     rgba = scalarMap.to_rgba(poles.AgeNominal[n])
#     hex_color = colors.rgb2hex(rgba)
    
color_1 = scalarMap.to_rgba(1)
color_1_hex = colors.rgb2hex(color_1)
color_5_6 = scalarMap.to_rgba(5.6)
color_5_6_hex = colors.rgb2hex(color_5_6)
color_10 = scalarMap.to_rgba(10)
color_10_hex = colors.rgb2hex(color_10)
color_31 = scalarMap.to_rgba(31.6)
color_31_hex = colors.rgb2hex(color_31)
color_99 = scalarMap.to_rgba(99)
color_99_hex = colors.rgb2hex(color_99)
color_997 = scalarMap.to_rgba(997.3)
color_997_hex = colors.rgb2hex(color_997)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,3,1)
plt.plot(BRIC_20c_raw['T [K]'],BRIC_20c_raw["X'(99.9Hz)"],color=color_99_hex,label="$\chi$' 99.9 Hz")
plt.plot(BRIC_20c_raw['T [K]'],BRIC_20c_raw["X'(10.0Hz)"],color=color_10_hex,label="$\chi$' 10.0 Hz")
plt.plot(BRIC_20c_raw['T [K]'],BRIC_20c_raw["X'(1.0Hz)"],color=color_1_hex,label="$\chi$' 1.0 Hz")
plt.ylabel('$\chi$ (m$^3$/kg)')
plt.xlabel('temperature (K)')
#plt.title('in-phase')
plt.legend()
plt.subplot(2,3,2)
plt.plot(BRIC_20c_modeled['T [K]'],BRIC_20c_modeled["X'(99.9Hz)"],color=color_99_hex,label="$\chi$' 99.9 Hz")
plt.plot(BRIC_20c_modeled['T [K]'],BRIC_20c_modeled["X'(10.0Hz)"],color=color_10_hex,label="$\chi$' 10.0 Hz")
plt.plot(BRIC_20c_modeled['T [K]'],BRIC_20c_modeled["X'(1.0Hz)"],color=color_1_hex,label="$\chi$' 1.0 Hz")
plt.ylabel('$\chi$ (m$^3$/kg)')
plt.xlabel('temperature (K)')
#plt.title('in-phase (paramagnetism removed)')
plt.legend()
plt.subplot(2,3,3)
plt.plot(BRIC_20c_raw['T [K]'],BRIC_20c_raw["X''(99.9Hz)"],color=color_99_hex,label='$\chi$" 99.9 Hz')
plt.plot(BRIC_20c_raw['T [K]'],BRIC_20c_raw["X''(10.0Hz)"],color=color_10_hex,label='$\chi$" 10.0 Hz')
plt.plot(BRIC_20c_raw['T [K]'],BRIC_20c_raw["X''(1.0Hz)"],color=color_1_hex,label='$\chi$" 1.0 Hz')
plt.plot(BRIC_20c_raw['T [K]'],BRIC_20c_raw["-pi/2 dX/dln(f)(10.0Hz)"],linestyle='dotted',color='black',label='-$\pi$/2 d$\chi$/dln(f)\n(10.0Hz)')
plt.ylabel('$\chi$ (m$^3$/kg)')
plt.xlabel('temperature (K)')
ymin, ymax = plt.ylim()
plt.ylim(0,ymax)
#plt.title('out-of-phase')
plt.legend()

plt.subplot(2,3,4)
plt.plot(BRIC_31c_raw['T [K]'],BRIC_31c_raw["X'(99.9Hz)"],color=color_99_hex,label="$\chi$' 99.9 Hz")
plt.plot(BRIC_31c_raw['T [K]'],BRIC_31c_raw["X'(10.0Hz)"],color=color_10_hex,label="$\chi$' 10.0 Hz")
plt.plot(BRIC_31c_raw['T [K]'],BRIC_31c_raw["X'(1.0Hz)"],color=color_1_hex,label="$\chi$' 1.0 Hz")
plt.ylabel('$\chi$ (m$^3$/kg)')
plt.xlabel('temperature (K)')
#plt.title('in-phase')
plt.legend()
plt.subplot(2,3,5)
plt.plot(BRIC_31c_modeled['T [K]'],BRIC_31c_modeled["X'(99.9Hz)"],color=color_99_hex,label="$\chi$' 99.9 Hz")
plt.plot(BRIC_31c_modeled['T [K]'],BRIC_31c_modeled["X'(10.0Hz)"],color=color_10_hex,label="$\chi$' 10.0 Hz")
plt.plot(BRIC_31c_modeled['T [K]'],BRIC_31c_modeled["X'(1.0Hz)"],color=color_1_hex,label="$\chi$' 1.0 Hz")
plt.ylabel('$\chi$ (m$^3$/kg)')
plt.xlabel('temperature (K)')
#plt.title('in-phase (paramagnetism removed)')
plt.legend()
plt.subplot(2,3,6)
plt.plot(BRIC_31c_raw['T [K]'],BRIC_31c_raw["X''(99.9Hz)"],color=color_99_hex,label='$\chi$" 99.9 Hz')
plt.plot(BRIC_31c_raw['T [K]'],BRIC_31c_raw["X''(10.0Hz)"],color=color_10_hex,label='$\chi$" 10.0 Hz')
plt.plot(BRIC_31c_raw['T [K]'],BRIC_31c_raw["X''(1.0Hz)"],color=color_1_hex,label='$\chi$" 1.0 Hz')
plt.plot(BRIC_31c_raw['T [K]'],BRIC_31c_raw["-pi/2 dX/dln(f)(10.0Hz)"],linestyle='dotted',color='black',label='-$\pi$/2 d$\chi$/dln(f)\n(10.0Hz)')
plt.ylabel('$\chi$ (m$^3$/kg)')
plt.xlabel('temperature (K)')
ymin, ymax = plt.ylim()
plt.ylim(0,ymax)
#plt.title('out-of-phase')
plt.legend()
plt.tight_layout()
```
